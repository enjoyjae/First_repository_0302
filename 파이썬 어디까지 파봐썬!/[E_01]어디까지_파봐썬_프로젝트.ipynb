## **10-3. 프로젝트 2 : 리스트 평탄화**

```python
def flatten(data):
    output = []  # 빈 리스트를 만듭니다.
    for item in data :
        if type(item) == list:
            output += flatten(item)
        else:
            output.append(item)
    return output
```

```python
# 결과_Test
a = [[1, 2], 3, [[4, 5, 6], 7], 8, 9]
flatten(a)
```

## **10-4. 프로젝트 3 : 10 이하 숫자만 곱해주는 함수**

```python
def mul(*values):
    output = 1
    for num in values:
        if num <= 10:
            output *= num
    return output
```

```python
# 결과_Test_1
print(f'Test_1:',mul(2, 3, 4, 5))

# 결과_Test_2
print(f'Test_2:',mul(3, 12, 10))
```
